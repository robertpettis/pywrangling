## Cell tag demos

The following cells demonstrate slide-only display tags.
These are enabled by `apply_beamer_theme(..., cell_tags=True)` or by calling `apply_cell_tags()` directly.

**Slide-only tags** (cells stay fully editable in notebook view):
- `slide_remove_input` — hide code input during slideshow only
- `slide_remove_output` — hide output during slideshow only
- `slide_side_by_side` — code & output side by side during slideshow only
- `slide_hide_prompt` — hide In[N]/Out[N] prompts during slideshow only

**Always-active tags** (hidden in notebook *and* slideshow):
- `remove_input`, `remove_output`, `side_by_side`, `hide_prompt`

````python
# Tag: slide_remove_input — code is hidden during slideshow only
# (you can still see and edit this cell in the notebook!)
import pandas as pd
pd.DataFrame({
    "Planet": ["Mercury", "Venus", "Earth", "Mars"],
    "Moons": [0, 0, 1, 2],
    "Orbital Period (days)": [88, 225, 365, 687],
})
````

````python
# Tag: slide_remove_output — output is hidden during slideshow only
# (you can still see and edit this cell in the notebook!)
def fibonacci(n):
    """Return the first n Fibonacci numbers."""
    seq = [0, 1]
    for i in range(2, n):
        seq.append(seq[-1] + seq[-2])
    return seq[:n]

print(fibonacci(10))
````

````python
# ============================================================
# Automated test: generate all themes and verify files exist
# ============================================================
import os
from pywrangling.jupyter_slides_functions import create_beamer_rise_theme

themes = ["copenhagen", "madrid", "warsaw", "annarbor", "berkeley", "stanford", "gamecocks"]
results = {}

for t in themes:
    try:
        path = create_beamer_rise_theme(theme=t, skip_font_download=True)
        css_exists = os.path.exists(path)
        theme_file = f"beamer_assets/themes/beamer_{t}.css"
        theme_exists = os.path.exists(theme_file)
        results[t] = {
            "rise.css": css_exists,
            "theme_css": theme_exists,
        }
        print(f"  {t:12s}  rise.css={css_exists}  theme_css={theme_exists}  OK")
    except Exception as e:
        results[t] = {"error": str(e)}
        print(f"  {t:12s}  ERROR: {e}")

# Verify aliases work
try:
    create_beamer_rise_theme(theme="ann arbor", skip_font_download=True)
    print(f"  {'ann arbor':12s}  alias works  OK")
except Exception as e:
    print(f"  {'ann arbor':12s}  alias ERROR: {e}")

try:
    create_beamer_rise_theme(theme="south carolina", skip_font_download=True)
    print(f"  {'south carolina':14s}  alias works  OK")
except Exception as e:
    print(f"  {'south carolina':14s}  alias ERROR: {e}")


# Verify invalid theme raises
#try:
#    create_beamer_rise_theme(theme="nonexistent", skip_font_download=True)
#    print(f"  {'nonexistent':12s}  should have raised ValueError!  FAIL")
#except ValueError:t
#    print(f"  {'nonexistent':12s}  ValueError raised correctly  OK")

# Verify apply_beamer_theme exists and is callable
from pywrangling.jupyter_slides_functions import apply_beamer_theme
print(f"  apply_beamer_theme callable: {callable(apply_beamer_theme)}  OK")

print("\nAll tests passed!")
````
